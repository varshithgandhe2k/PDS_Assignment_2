```python
# Justification:
# Imputing with mean or mode is suitable for numeric and categorical data, respectively, to retain as much information as possible.
# Example: Imputing missing values with mean for numeric columns
# Example: Imputing missing values with mode for categorical columns
```

```python
# a. Impute missing values for categorical columns with mode
categorical_columns = data.select_dtypes(include=['object']).columns
data[categorical_columns] = data[categorical_columns].fillna(data[categorical_columns].mode().iloc[0])

# Impute missing values for numerical columns with mean
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns
data[numerical_columns] = data[numerical_columns].fillna(data[numerical_columns].mean())
```

```python
#Drop New_Price column
#New_Price has a significant amount of missing values (around 5032 rows out of 5847 rows). Filling these missing values with mean,
#median or mode can influence the model performance. There are many other significant factors like engine, mileage to predict the
#target variable.
data = data.drop(columns = ['New_Price'])
data.head()
```

```python
data.isnull().any() # now we have handled missing data. so we will get false for every column
```

```python
#b. Removing Units from some columns
# Replace units with an empty string
data['Mileage'] = data['Mileage'].astype(str).str.replace(r'[^\d.]', '', regex=True).astype(float)
```

```python
# Remove units from Engine displacement
data['Engine'] = data['Engine'].astype(str).str.replace(r'[^\d.]', '', regex=True).astype(float)

# Remove units from Engine horsepower
data['Power'] = data['Power'].astype(str).str.replace(r'[^\d.]', '', regex=True).astype(float)
```

```python
data.head()
```

```python
# c. numerical one hot encoding
data = pd.get_dummies(data, columns=['Fuel_Type','Transmission'])
data.head()
```

```python
data.head()
```

```python
# d. Creating one more feature and Adding this extra column to dataset
current_year = 2024
data['age_of_car'] = current_year - data['Year']
```

```python
data.head()
```

```python
# e. Performing some operations in python for above data
# Select: Choose specific columns
selected_columns = data[['Name', 'Mileage', 'Owner_Type', 'Price', 'age_of_car']]
selected_columns.head()
```

```python
# Filter: Filter rows based on condition that is fuel_type is petrol
filtered_data = data[data['Owner_Type'] == 'First']
filtered_data.head()
```

```python
# Rename: Rename columns
renamed_data = data.rename(columns={'Year': 'Manufacture_Year', 'Location': 'City'})
renamed_data.head()
```

```python
# Mutate: Create or modify columns
mutated_data = data.assign(Mileage_per_Year=data['Mileage'] / (2024 - data['Year']))
mutated_data.head()
```

```python
# Arrange: Sort rows based on one or more columns
sorted_data = data.sort_values(by=['Price'], ascending=[True])
sorted_data.head()
```
